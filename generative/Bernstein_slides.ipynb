## Demonstration of the Bernstein polynomials

$$
B_{k}^{M}(z) = \binom{M}{k} z^k (1 - z)^{M-k}
$$

```python
import scipy.special
import numpy as np
import matplotlib.pyplot as plt

def bernstein_basis(n, k, z):
    """
    Compute the Bernstein basis polynomial of degree n and order k at point z.

    Parameters:
    n (int): Degree of the polynomial.
    k (int): Order of the basis function.
    z (float): Point at which to evaluate the polynomial.

    Returns:
    float: Value of the Bernstein basis polynomial at point z.
    """
    return scipy.special.comb(n, k) * (z ** k) * ((1 - z) ** (n - k))

# Degree of the polynomial
n = 4

# Values of z at which to evaluate the basis functions
z = np.linspace(0, 1, 100)

# Plot the Bernstein basis polynomials for n = 4
plt.figure(figsize=(10, 6))
for k in range(n + 1):
    plt.plot(z, bernstein_basis(n, k, z), label=r'$B_{}^M{}(z)$'.format(k, n))


#Ensure a square aspect ratio
plt.gca().set_aspect('equal', adjustable='box')
plt.title(f'Bernstein Basis Polynomials of Order {n}')
plt.xlabel('z')
plt.ylabel('Basis Value')
plt.legend()
plt.grid(True)
plt.show()
```

g(z) = \sum_{k=0}^{M} \vartheta_k B_{k}^{M}(z) = \sum_{k=0}^{M} \vartheta_k \binom{M}{k} z^k (1 - z)^{M-k}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
import scipy.special

def bernstein_basis(M, k, z):
    return scipy.special.comb(M, k) * (z ** k) * ((1 - z) ** (M - k))

def plot_bernstein_combination(v0, v1, v2, v3, v4):
    z = np.linspace(0, 1, 100)
    M = 4  # Degree of the polynomial
    coefficients = [v0, v1, v2, v3, v4]
    combination = sum(v * bernstein_basis(M, k, z) for k, v in enumerate(coefficients))

    plt.figure(figsize=(10, 6))
    plt.plot(z, combination, label='Combination')
    plt.title('Combination of Bernstein Basis Polynomials')
    plt.xlabel('z')
    plt.ylabel('Basis Value')
    plt.grid(True)
    plt.legend()
    plt.show()

# Interactive widget with named coefficients
interact(plot_bernstein_combination, 
         v0=(0, 5, 0.1), 
         v1=(0, 5, 0.1), 
         v2=(0, 5, 0.1), 
         v3=(0, 5, 0.1), 
         v4=(0, 5, 0.1))
```
