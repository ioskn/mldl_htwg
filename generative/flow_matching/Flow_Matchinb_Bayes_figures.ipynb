```python
from graphviz import Digraph
dot = Digraph(format="png")
dot.attr("node", shape="box", style="rounded", color="darkblue", penwidth="2")
dot.node("NN", "NN", shape="box", style="rounded")
dot.attr("node", shape="plaintext")
dot.node("x", "x")
dot.node("p", "p(y|x)")
dot.edge("x", "NN", label="", color="darkblue")
dot.edge("NN", "p", label="", color="darkblue")
dot.attr(rankdir="LR")
dot.graph_attr["dpi"] = "300"  # ✅ Correct
dot.render("prob_regression", format="png")
dot
```

```python
from graphviz import Digraph

dot = Digraph(format="png")

dot.attr(rankdir="LR")

dot.attr("node", shape="box", style="rounded", color="darkblue", penwidth="2")

dot.node("NN", "NN", shape="box", style="rounded")

dot.attr("node", shape="plaintext")
dot.node("D", "D")
dot.node("p", "<p(&theta;|D)>")

dot.edge("D", "NN", label="", color="darkblue")
dot.edge("NN", "p", label="", color="darkblue")
dot.graph_attr["dpi"] = "300"  # ✅ Correct
dot.render("sbi", format="png", renderer="cairo")
dot
```
