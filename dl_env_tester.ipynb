# Testing the Deep Learning Environment

Testing for the Deep Learning Environment with Keras and **pytorch** backend

### Testing the data science stack



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import tqdm

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.show()

pd.DataFrame({"x": x, "y": y}).head()
import time
for i in tqdm.tqdm(range(10)):
    time.sleep(0.05)
```

### Testing the dl-stack

```python
import os
os.environ["KERAS_BACKEND"] = "torch"

import keras
print("Keras:", keras.__version__)
print("Backend:", keras.backend.backend())

import torch

# Print python version
print("PyTorch:", torch.__version__)
print("MPS available:", torch.backends.mps.is_available())

device = "mps" if torch.backends.mps.is_available() else "cpu"
x = torch.randn(1000, 1000, device=device)
y = x @ x
print("Result:", y.mean(), "on", device)


print("Keras:", keras.__version__)
print("Backend:", keras.backend.backend())
```
